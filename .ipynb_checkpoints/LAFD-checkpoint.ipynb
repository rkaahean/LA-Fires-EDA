```python
df = pd.read_csv('LAFD_Response_Metrics_-_Raw_Data.csv')
```

Creating acronyms

```python
ID = 'Randomized Incident Number'
FID = 'First In District'
DISP_SEQ = 'Dispatch Sequence'
STAT = 'Dispatch Status'
TYPE = 'Unit Type'
LEVEL = 'PPE Level'
ICD_CREAT_TIME = 'Incident Creation Time (GMT)'
DISP_TIME = 'Time of Dispatch (GMT)'
LEAVE_TIME = 'En Route Time (GMT)'
REACH_TIME = 'On Scene Time (GMT)'

CPS = 'Call Processing Time'
CPS_DUR = 'Call Processing Duration (ms)'
TURN_DUR = 'Turnout Time (ms)'
TRAV_DUR = 'Travel Time (ms)'
```

#### Time data

##### Processing the metrics

```python
data[ICD_CREAT_TIME] = pd.to_datetime(data[ICD_CREAT_TIME], format = '%H:%M:%S.%f')
data[DISP_TIME] = pd.to_datetime(data[DISP_TIME], format = '%H:%M:%S.%f')
data[LEAVE_TIME] = pd.to_datetime(data[TRAV_TIME], format = '%H:%M:%S.%f')
data[REACH_TIME] = pd.to_datetime(data[REACH_TIME], format = '%H:%M:%S.%f')
```

Removing pd.NaT data

```python
data = data.loc[~np.isnat(data[LEAVE_TIME]), :]
data = data.loc[~np.isnat(data[REACH_TIME]), :]
```

Creating the variables mentioned in the FAQ

_Call processing time._ **(CPS)**

Note: 
    - It is possible that a fire breaks out in 1 district and then spreads to the other districts are the fire 
    grows.
    - However, within the same incident and the same district, the creation label can be of multiple 
    different values.
**For every unique ICT** per incident, compute CPS

_Turnout Time._ (**TURN_DUR**)

Straightforward:
    - LEAVE_TIME - DISP_TIME
    
_Travel Time._ (**TRAV_DUR**)
    - Like _CPS_, groupby ID and ICD_CREAT_TIME.
    - 1st unit that reached - 1st unit that left , for the given grouping.

```python
#Example
#The calls can keep on coming, for the same incident and the same district
data[data[ID] == 201301808922].head()
```

```python
group = data.groupby([ID, ICD_CREAT_TIME])[DISP_TIME].min()
data_index = data.set_index([ID, ICD_CREAT_TIME])

for row in tqdm(range(len(group.index))):
    data_index.at[group.index[row], CPS]  = group.values[row]
    
data = data_index.reset_index()
data[CPS_DUR] = (data[CPS] - data[ICD_CREAT_TIME]).astype(np.int64) / int(1e6)
```

A Side effect:
    - There may be -ve durations. min(23:40, 00:01) is 00:01, even though timewise its not true. 
        Since edge cases, easier to drop them (the single record, not the entire incident)

```python
data = data[data[CPS_DUR] > 0]
```

Turnout Time

```python
#If LEAVE_TIME doesn't exist, assign pd.NaT
data[TURN_DUR] = (data[LEAVE_TIME] - data[DISP_TIME]).astype(np.int64) / int(1e6)
data = data[data[TURN_DUR] > 0]

data_index = data.set_index([ID, ICD_CREAT_TIME])
```

Travel Time

```python
trav_group = data.groupby([ID, ICD_CREAT_TIME])
first_leave = trav_group[LEAVE_TIME].min()
first_reach = trav_group[REACH_TIME].min()

for row in tqdm(range(len(first_leave.index))):
    data_index.at[first_leave.index[row], TRAV_DUR] = \
    (first_reach.values[row] - first_leave.values[row]).astype(np.int64) / int(1e6)
```

Removing anomalies

```python
data = data_index.reset_index()
data = data[data[TRAV_DUR] > 0]
```

Saving this sample dataset.

```python
data.to_csv('metric_data.csv')
```
